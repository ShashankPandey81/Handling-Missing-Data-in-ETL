Q1. What are the most common reasons for missing data in ETL pipelines?

- Missing data can occur for several reasons during the ETL process. Sometimes users leave certain fields blank while filling out forms. Data may also be lost because of system errors, network failures, or problems during data transfer. In some cases, different data sources may not contain the same information, which leads to missing values after integration. Human errors during data entry are also a common reason for missing data.

Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?

- Deleting rows with missing values without checking them first can lead to the loss of useful information. If many records are removed, the dataset becomes smaller and may no longer represent the actual population. This can reduce the accuracy of reports and machine learning models. It is always better to understand why the data is missing before deciding whether to remove or replace it.

Q3. Explain the difference between Listwise deletion and Column deletion. Also mention one scenario where each is appropriate.

- Listwise deletion means removing the entire row if it contains one or more missing values. This method is suitable when only a few rows have missing data and removing them will not affect the overall analysis.

Example:
If only 2 out of 1,000 customer records have missing values, those rows can be safely removed.

Column deletion means removing an entire column when a large portion of its values is missing or when the column is not important for the analysis.

Example:
If a "Middle Name" column has 90% missing values and is not required for the project, it can be deleted.

Q4. Why is median imputation preferred over mean imputation for skewed data such as income?

- The median is preferred because it is less affected by extremely high or low values. Income data is usually skewed, with a few people earning much more than the rest. These extreme values can increase the mean and make it less representative. The median represents the middle value, so it gives a better estimate when the data is not evenly distributed.

Q5. What is forward fill and in what type of dataset is it most useful?

- Forward fill is a technique where a missing value is replaced with the last available value in the dataset. It is mainly used in time-series data where observations are recorded over time.

For example, if the stock price for one day is missing, the previous day's price can be used until a new value becomes available. This method is commonly used in financial, weather, and sensor datasets.

Q6. Why should flagging missing values be done before imputation in an ETL workflow?

- Flagging missing values means creating a separate indicator that shows which values were originally missing. This should be done before replacing the missing values because it keeps a record of the original data. Later, analysts or machine learning models can use this information to understand whether the missing values have any impact on the results. It also improves transparency and makes the data preparation process easier to track.

Q7. Consider a scenario where income is missing for many customers. How can this missingness itself provide business insights?

- If income information is missing for many customers, it may indicate a pattern rather than just a data quality issue. For example, customers with higher incomes might choose not to share their financial details because of privacy concerns. It could also mean that a particular branch, website, or data collection process is not asking for income consistently.

By studying this missing data, a company can identify problems in the data collection process, improve customer forms, and better understand customer behavior. In this way, missing values themselves can provide useful business insights instead of being treated only as errors.
